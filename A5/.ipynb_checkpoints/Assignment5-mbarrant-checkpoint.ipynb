```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#1

def dot_product(vector1, vector2):
    dot_product = np.vdot(vector1, vector2)  # Computes the dot product of complex vectors
    magnitude = abs(dot_product)
    real_part = dot_product.real
    imaginary_part = dot_product.imag
    phase_angle = np.angle(dot_product)
    
    return {
        "dot_product": dot_product,
        "magnitude": magnitude,
        "real_part": real_part,
        "imaginary_part": imaginary_part,
        "phase_angle": phase_angle
    }

# Test
vector1 = np.array([1 - 4j, 5 - 4j, 5 + 3j, 5 + 3j])
vector2 = np.array([1 + 1j, 3, 5 + 3j, 5 - 3j])

result = dot_product(vector1, vector2)
print(result)
```

## Explanation of the Solution

To solve this problem, I created a Python function that computes the dot product of two vectors of complex numbers and extracts key components from the result.

### **Step 1: Compute the Dot Product**
used `numpy.vdot(vector1, vector2)` to compute the dot product of the two complex vectors. The `vdot` function correctly handles complex conjugation in dot product calculations.

### **Step 2: Extract Key Components**
Once the dot product was computed, then extracted the following components:

- **Magnitude:** Calculated using `abs(dot_product)`, which gives the modulus of the complex number.
- **Real Part:** Accessed using `dot_product.real`, which returns the real component.
- **Imaginary Part:** Accessed using `dot_product.imag`, which returns the imaginary component.
- **Phase Angle:** Computed using `numpy.angle(dot_product)`, which returns the angle (in radians) of the complex number in the polar coordinate system.

### **Step 3: Returning the Results**
The function returns a dictionary containing:
- The **dot product**
- Its **magnitude**
- Its **real part**
- Its **imaginary part**
- Its **phase angle**

### **Testing**
function with two vectors of complex numbers, following the problem statement, and printed the results to verify correctness.

```python
#2

def compute_dtft(signal, time_samples, num_freqs=1000):
   
    N = len(time_samples)
    freq_range = np.linspace(-10, 10, num_freqs)  # Frequency range for analysis
    dtft = np.zeros(num_freqs, dtype=complex)

    for i, f in enumerate(freq_range):
        dtft[i] = np.sum(signal * np.exp(-1j * 2 * np.pi * f * time_samples))  # DTFT formula

    return freq_range, np.abs(dtft)

# Define time samples
t = np.linspace(0, 1, 1000, endpoint=False)  # 1 second duration, 1000 samples

# Define signals
signal_1 = 6 * np.sin(2 * np.pi * 3 * t) + 3 * np.sin(2 * np.pi * 6 * t)
signal_2 = np.sin(2 * np.pi * 3 * t) + np.cos(2 * np.pi * 6 * t)

print("Signal 1:")
print(signal_1)

print("Signal 2:")
print(signal_2)

# Compute DTFT
freqs_1, dtft_1 = compute_dtft(signal_1, t)
freqs_2, dtft_2 = compute_dtft(signal_2, t)

# Plot results
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(freqs_1, dtft_1)
plt.title("DTFT of 6sin(23t) + 3sin(26t)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")

plt.subplot(1, 2, 2)
plt.plot(freqs_2, dtft_2)
plt.title("DTFT of sin(23t) + cos(26t)")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Magnitude")

plt.tight_layout()
plt.show()
```

## DTFT Computation Explanation

implemented a **Discrete-Time Fourier Transform (DTFT)** function to identify the frequency components of the given signals.

### **Method:**
- The DTFT formula:  
  \[
  X(f) = \sum x[n] e^{-j 2\pi f n}
  \]
- We evaluated this sum over a range of frequencies to find dominant components.

### **Applying to Given Signals:**
1. **\(6 \sin(2\pi 3t) + 3 \sin(2\pi 6t)\)**  Peaks at **3 Hz & 6 Hz**  
2. **\(\sin(2\pi 3t) + \cos(2\pi 6t)\)**  Peaks at **3 Hz & 6 Hz**, different amplitudes due to cosine.  

### **Conclusion:**
The DTFT correctly identifies the **dominant frequencies**, confirming the signal's basis functions.



```python
# 3
from scipy.fftpack import fft2, ifft2, fftshift, ifftshift
from skimage import color, data

# Load astronaut image and convert to grayscale
astronaut = data.astronaut()
gray_image = color.rgb2gray(astronaut)

fft_image = fft2(gray_image)
fft_shifted = fftshift(fft_image)

# Compute amplitude and phase
amplitude = np.abs(fft_shifted)
phase = np.angle(fft_shifted)

# Reconstruct image using inverse FFT
reconstructed_image = np.real(ifft2(ifftshift(fft_shifted)))

# Compute reconstruction error
error = np.abs(gray_image - reconstructed_image)

# Remove high frequencies by zeroing out outer regions
rows, cols = gray_image.shape
crow, ccol = rows // 2, cols // 2 
cutoff = 30  

filtered_fft = fft_shifted.copy()
filtered_fft[crow - cutoff:crow + cutoff, ccol - cutoff:ccol + cutoff] = 0
filtered_image = np.real(ifft2(ifftshift(filtered_fft)))

# Plot original
fig, ax = plt.subplots(1, 4, figsize=(15, 5))
ax[0].imshow(gray_image, cmap="gray")
ax[0].set_title("Original Grayscale Image")
ax[1].imshow(np.log(1 + amplitude), cmap="gray")
ax[1].set_title("Log Amplitude Spectrum")
ax[2].imshow(reconstructed_image, cmap="gray")
ax[2].set_title("Reconstructed Image")
ax[3].imshow(filtered_image, cmap="gray")
ax[3].set_title("High Frequencies Removed")

for a in ax:
    a.axis("off")

plt.show()
```

## Fourier Transform on Astronaut Image  

### **Steps:**
1. **Load Image & Convert to Grayscale:**  
   - Used `matplotlib.image.imread()` to load the astronaut image.  
   - Converted it to grayscale using `rgb2gray()`.  

2. **Compute Fourier Transform:**  
   - Applied `scipy.fftpack.fft2()` to obtain frequency components.  
   - Used `fftshift()` to center low frequencies.  

3. **Extract Amplitude & Phase:**  
   - **Amplitude:** `np.abs(fft_image)`  
   - **Phase:** `np.angle(fft_image)`

4. **Reconstruct Image & Calculate Error:**  
   - Used `ifft2()` to reconstruct from full frequency data.  
   - Computed **error** as the difference between original and reconstructed images.  

5. **Remove High Frequencies & Reconstruct:**  
   - Set high-frequency components to zero.  
   - Used `ifft2()` to reconstruct a smoother image.  

### **Conclusion:**
- The Fourier transform separates image details into amplitude (intensity) and phase (structure).  
- Removing high frequencies blurs the image, showing their importance in fine details.  


```python
# 4
from skimage import color, data

# Load astronaut image and convert to grayscale
astronaut = data.astronaut()
gray_image = color.rgb2gray(astronaut)

# Compute 2D Fourier Transform and shift zero-frequency component to the center
fft_image = fft2(gray_image)
fft_shifted = fftshift(fft_image)

# Generate a Gaussian distribution centered in the frequency domain
rows, cols = gray_image.shape
x, y = np.meshgrid(np.linspace(-1, 1, cols), np.linspace(-1, 1, rows))
sigma = 0.2  # Standard deviation of the Gaussian
gaussian_filter = np.exp(- (x**2 + y**2) / (2 * sigma**2))

# Multiply element-wise with the Fourier coefficients
filtered_fft = fft_shifted * gaussian_filter

filtered_image = np.real(ifft2(ifftshift(filtered_fft)))

# Plot
fig, ax = plt.subplots(1, 3, figsize=(15, 5))
ax[0].imshow(gray_image, cmap="gray")
ax[0].set_title("Original Grayscale Image")
ax[1].imshow(gaussian_filter, cmap="gray")
ax[1].set_title("Gaussian Filter")
ax[2].imshow(filtered_image, cmap="gray")
ax[2].set_title("Transformed Image (Filtered)")

for a in ax:
    a.axis("off")

plt.show()
```
